# 1.Bike Sales Dataset Specification

a) Sales Volume

b) Sales Costs

### I recommend a change in the data type:

Change:

**Date** *object* to *date*

**Customer_Gender** *object* to *string*

# 5.Sales costs, profit, and revenue:

### Create plot showing annual costs, profit, and revenue.



### Data Preparation

1) Create new dataset containing the columns Year, Costs, Profit, Revenue

2) Exclude 2016 data, because the 2016 data is only till 2016-07-31

3) Aggrigate the data by Year column

```python
df_annual_data = pd.DataFrame(df_sales, columns=["Year", "Cost", "Profit", "Revenue"])
df_annual_data.head()
```

```python
# Exclude Year 2016

flt = (df_sales["Year"] < 2016)

# Annual dtaset update: Year 2016 excluded
df_annual_data = df_annual_data.loc[flt, :]

df_annual_data.head()
```

```python
# Grouping Year
df_annual_agregation = df_annual_data.groupby("Year")

# Aggregation Costs, Profit, Revenue using sum()
df_plot_data = df_annual_agregation.agg("sum")

# Displaying Plot Data
display(df_plot_data)
```

### Making a Plot: Annual Costs, Profit, Revenue

1) Area Chart

2) Line Charts showing Costs, Profit, Revenue by year

```python
import matplotlib.pyplot as plt
# Ploting data

year = df_plot_data.index.values
annual_cost = df_plot_data["Cost"]/1000
annual_profit = df_plot_data["Profit"]/1000
annual_revenue = df_plot_data["Revenue"]/1000


# Plot Colors
colors = {
    "annual_cost": "#FEAE73",
    "annual_profit": "#006955",
    "annual_revenue": "#DF0113"
}


plt.figure(figsize = (10, 10))

# Linear Chart
plt.subplot(211)
plt.plot(year, annual_cost, label="Annual Cost", color=colors["annual_cost"], linewidth=2)
plt.plot(year, annual_profit, label="Annual Profit", color=colors["annual_profit"], linewidth=2)
plt.plot(year, annual_revenue, label="Annual Revenue", color=colors["annual_revenue"], linewidth=2)

# Add Grid
plt.grid()
# Update x-axis ticks
plt.xticks(year, fontsize=12)
# Update y-axis ticks
plt.yticks(fontsize=12)
# Update y-axis label
plt.ylabel("Amount in 10^3 USD", fontsize=12)
# Show Line Chart Legend
plt.legend()

# Area Chart
plt.subplot(212)
plt.stackplot(year, annual_cost, annual_profit,
             labels = ["Annual Cost", "Annual Profit"],
             colors = [colors["annual_cost"], colors["annual_profit"]]
             )

plt.plot(year, annual_revenue, label="Annual Revenue", color=colors["annual_revenue"], linewidth=3)

# Show Line Chart Legend
plt.legend()

# Add Grid
plt.grid()

# Update x-axis ticks
plt.xticks(year, fontsize=12)

# Update y-axis ticks
plt.yticks(fontsize=12)

# Update x-axis label
plt.xlabel("Year", fontsize=12)

# Update y-axis label
plt.ylabel("Amount in 10^3 USD", fontsize=12)

# Plot Title
plt.suptitle("\nAnnual Costs, Profit, Revenue \n(in thousand USD)", fontsize=18)

# Display plot
plt.show()

```

### Create plot showing monthly costs, profit, and revenue.

```python
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter

df_annual_month = pd.DataFrame(df_sales, columns=["Month", "Cost", "Profit", "Revenue"])
df_annual_month = df_annual_month.set_index('Month')
# Grouping Month
df_annual_agregation_month = df_annual_month.groupby("Month")

# Aggregation Costs, Profit, Revenue using sum()
df_plot_data_month = df_annual_agregation_month.agg("sum")

# Displaying Plot Data
display(df_plot_data_month)

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
# rearrange the rows in the DataFrame
df_plot_data_month = df_plot_data_month.reindex(month_order)
display(df_plot_data_month)

# Plot
ax = df_plot_data_month.plot(kind='bar', figsize=(20, 10))
ax.set_xlabel('Month', fontsize=15)
ax.set_ylabel('Amount, USD', fontsize=15)
ax.set_title('Cost, Profit, and Revenue by Month', fontsize=20)

# add a legend
plt.legend(loc='upper left', ncols=3, fontsize=15)

# Setting axis signatures
ax.tick_params(axis = 'x', labelrotation=45, labelsize=15)
ax.tick_params(axis = 'y',labelsize=15)
sf = ScalarFormatter()
sf.set_powerlimits((-7, 7))
ax.yaxis.set_major_formatter(sf)

# display a graph
plt.show()
```

```python
def addlabels(x, y):
    font = {
        'family': 'arial',
        'color':  'black',
        'weight': 'bold',
        'size': 16,
        }

    for i in range(len(x)):
        plt.text(i, y[i], round(y[i], 2), fontdict=font, ha = 'center')
```

```python
import matplotlib.pyplot as plt
# Calculate monthly totals for costs, revenue, and profit

print(df_plot_data_month)
month = df_plot_data_month.index.values
annual_cost = df_plot_data_month["Cost"]/1000000
annual_profit = df_plot_data_month["Profit"]/1000000
annual_revenue = df_plot_data_month["Revenue"]/1000000

# Plot Colors
colors = {
    "annual_cost": "#FEAE73",
    "annual_profit": "#006955",
    "annual_revenue": "#DF0113"
}

# Plot
plt.figure(figsize = (20, 17))

# Linear Chart
ax1 = plt.subplot(211)
plt.plot(month, annual_cost, '-o', label="Annual Cost", color=colors["annual_cost"], linewidth=2)
addlabels(month, annual_cost)
plt.plot(month, annual_profit, '-o', label="Annual Profit", color=colors["annual_profit"], linewidth=2)
addlabels(month, annual_profit)
plt.plot(month, annual_revenue, '-o', label="Annual Revenue", color=colors["annual_revenue"], linewidth=2)
addlabels(month, annual_revenue)
# Add Grid
plt.grid()
# Update x-axis ticks
plt.xticks(month_order, fontsize=15)
# Update y-axis ticks
plt.yticks(fontsize=15)
# Update y-axis label
plt.ylabel("Amount in mln USD", fontsize=15)
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
# Show Line Chart Legend
plt.legend(loc='upper left', ncols=1, fontsize=15)


# Area Chart
ax2 = plt.subplot(212)
plt.stackplot(month, annual_cost, annual_profit,
             labels = ["Annual Cost", "Annual Profit"],
             colors = [colors["annual_cost"], colors["annual_profit"]]
             )
plt.plot(month, annual_revenue, label="Annual Revenue", color=colors["annual_revenue"], linewidth=3)
# Show Line Chart Legend
plt.legend(loc='upper left', ncols=1, fontsize=15)
# Add Grid
plt.grid()
# Update x-axis ticks
plt.xticks(month_order, fontsize=15)
# Update y-axis ticks
plt.yticks(fontsize=15)
# Update x-axis label
plt.xlabel("Month", fontsize=15)
# Update y-axis label
plt.ylabel("Amount in mln USD", fontsize=15)
# Plot Title
plt.suptitle("\nAnnual Costs, Profit, Revenue \n(in mln USD)", fontsize=20)
ax2.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
# Display plot
plt.show()
```

# 6.Customers by Age Group

### Create plot showing bike store customers by age group by year.



```python
plt.figure(figsize = (10, 10))

# Group the data by age group and year, and count the number of customers
df_customers = df_sales.groupby(['Age_Group', 'Year']).size().reset_index(name='count')

# Pivot the data so that each age group is a separate column
df_customers_pivot = df_customers.pivot(index='Year', columns='Age_Group', values='count')
print(df_customers_pivot)

# Create a stacked bar chart
ax1 = df_customers_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set the chart title and axis labels
ax1.set_title('Number of Customers by Age Group and Year')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Customers')
ax1.tick_params(axis = 'x', labelrotation=0)


# Create a stacked bar chart
ax2 = df_customers_pivot.plot(kind='bar', stacked=False, figsize=(10, 6))

# Set the chart title and axis labels
ax2.set_title('Number of Customers by Age Group and Year')
ax2.set_xlabel('Year')
ax2.set_ylabel('Number of Customers')
ax2.tick_params(axis = 'x', labelrotation=0)

plt.show()
```

### Create plot showing sales costs, profit, and revenue by year by age group.

```python
# df_data = pd.DataFrame(df_sales, columns=["Year", "Age_Group", 'Customer_Gender', "Cost", "Profit", "Revenue"])
df_data =df_sales.groupby(["Year", "Age_Group", 'Customer_Gender']).agg({'Cost': 'sum', 'Profit': 'sum', 'Revenue': 'sum'})

# Displaying Plot Data
display(df_data)
```

```python
# Group the data by age group, year, and the three metrics of interest, and sum the values
df_annuas = df_data.groupby(['Age_Group', 'Year'])[['Cost', 'Profit', 'Revenue']].sum().reset_index()

# Pivot the data so that each metric is a separate column
df_annuas_pivot = df_annuas.pivot(index='Year', columns='Age_Group')
print(df_annuas_pivot)

# Create chart 
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_size_inches(12, 10)  

df_annuas_pivot.plot(kind='bar', y='Cost',stacked=True, ax=ax1)
df_annuas_pivot.plot(kind='bar', y='Profit',stacked=True, ax=ax2)
df_annuas_pivot.plot(kind='bar', y='Revenue',stacked=True, ax=ax3)

# Set the chart title and axis labels
ax1.set_title('Cost by Year and Age Group')
ax2.set_title('Profit by Year and Age Group')
ax3.set_title('Revenue by Year and Age Group')

plt.xlabel('Year')
ax1.set_ylabel('Amount, USD')

ax1.tick_params(axis = 'x', labelrotation=0)
ax2.tick_params(axis = 'x', labelrotation=0)
ax3.tick_params(axis = 'x', labelrotation=0)

ax1.legend(loc='upper left', ncols=1)
ax2.legend(loc='upper left', ncols=1)
ax3.legend(loc='upper left', ncols=1)

plt.show()
```
